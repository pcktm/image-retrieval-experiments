```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('data/query_dog.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# generate ORB keypoints and descriptors
orb = cv2.ORB_create()
keypoints, descriptors = orb.detectAndCompute(gray, None)

# draw keypoints on top of the input image
image_with_keypoints_orb = cv2.drawKeypoints(gray, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(cv2.cvtColor(image_with_keypoints_orb, cv2.COLOR_BGR2RGB))
```

```python
# generate SIFT keypoints and descriptors
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(gray, None)

# draw keypoints on top of the input image
image_with_keypoints_sift = cv2.drawKeypoints(gray, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(cv2.cvtColor(image_with_keypoints_sift, cv2.COLOR_BGR2RGB))
```

```python
fast = cv2.FastFeatureDetector()
keypoints = fast.detect(gray, None)

image_with_keypoints_fast = cv2.drawKeypoints(gray, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(cv2.cvtColor(image_with_keypoints_fast, cv2.COLOR_BGR2RGB))
```
